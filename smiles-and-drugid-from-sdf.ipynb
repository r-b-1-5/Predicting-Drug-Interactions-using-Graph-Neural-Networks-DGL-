```python
from rdkit import Chem
```

```python
input_sdf = Chem.SDMolSupplier('/kaggle/input/drugbank-sdf/structures.sdf')
```

```python
print(len(input_sdf))
```

```python
molecules = [t for t in input_sdf if t is not None]
```

```python
print(len(molecules))
```

```python
drug_and_smiles_list = []
for mol in molecules:
    drugbank_id = mol.GetProp('DATABASE_ID')
#     print(drugbank_id)
    smiles = mol.GetProp('SMILES')
#     print(smiles)
    drug_and_smiles_list.append([drugbank_id, smiles])
#     break
```

```python
print(len(drug_and_smiles_list))
```

```python
# print(drug_and_smiles_list)
```

```python
import csv 
filename = "drugIDandSMILES.csv"

col_names = ['Drug ID', 'SMILES']
with open(filename, 'w') as csvfile123:
    csvwriter = csv.writer(csvfile123) 
        
    csvwriter.writerow(col_names) 
    
#     for interacting_pair in interaction_list:
#         print(interacting_pair)
#         break
#         csvwriter.writerow(interacting_pair)
    csvwriter.writerows(drug_and_smiles_list)
```
