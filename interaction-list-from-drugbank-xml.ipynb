```python
print(len(dict123))
drug_details = dict123['drugbank']['drug']
print(len(drug_details))
```

```python
import tqdm
interaction_list = []
for idx in range(len(drug_details)):
    drug_detail = drug_details[idx]
    try:
        drug_id = drug_detail['drugbank-id'][0]['#text']
    except:
        drug_id = drug_detail['drugbank-id']['#text']
#         print(drug_detail['drugbank-id']['#text'])
#         continue
#     print(drug_id)
    try:
        drug_interactions = drug_detail['drug-interactions']['drug-interaction']
    except: 
        drug_interactions = None
    if drug_interactions is None:
        continue
    for idx2 in range(len(drug_interactions)):
        try:
            interacting_drug = drug_interactions[idx2]
        except:
#             print(drug_id)
            interacting_drug = drug_interactions
#             print(drug_interactions)
#             break
        interacting_drug_id = interacting_drug['drugbank-id']
#         print(interacting_drug_id)
        interaction_list.append([drug_id, interacting_drug_id])
#         break
#     break    
```
